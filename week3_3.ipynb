```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#TASK №1

x = ["Джефф Безос", "Илон Маск","Бернар Арно","Билл Гейтс","Марк Цукерберг", "Уоррен Баффет","Ларри Эллисон","Ларри Пейдж"]
y = [177, 151, 150, 124, 97, 96, 93, 91.5]

fig, ax = plt.subplots()

ax.bar(x, y)

ax.set_facecolor('seashell')
fig.set_facecolor('floralwhite')
fig.set_figwidth(18)    #  ширина Figure
fig.set_figheight(6)    #  высота Figure

plt.show()
```

```python
# TASK №2 Количество заразившихся в России за всё время.

data = pd.read_csv('./04-27-2021.csv')
#print(data.head(6))

data2 = data[data.Country_Region=="Russia"]['Confirmed'].sum()

print("Количество заразившихся в России за все время ",data2)
```

```python
# TASK №3 Гистограмма заразившихся по регионам России.

data3 = data[['Country_Region','Province_State','Confirmed']]
data3 = data3[data3.Country_Region=="Russia"]
print(data3.head(7))

x_c = data3['Province_State'].tolist()
y_c = data3['Confirmed'].tolist()

fig, ax = plt.subplots()

ax.barh(x_c, y_c, color = 'red')
ax.set_yticks(x_c)
ax.set_yticklabels(x_c, rotation = 0)

ax.set_facecolor('seashell')
fig.set_facecolor('floralwhite')
fig.set_figwidth(10)    #  ширина Figure
fig.set_figheight(60)    #  высота Figure

plt.show()
```
